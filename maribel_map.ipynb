```python
import folium
import pandas as pd

# Read the CSV file into a DataFrame
df = pd.read_csv('muestreo.csv')
print(df)
```

```python
# Create a map centered at Viña del Mar
vdm_map = folium.Map(location=[-33.0245, -71.5528], zoom_start=13)

# Iterate through the DataFrame and add markers with names always shown
for index, row in df.iterrows():
    # Add circle marker
    """ circle_marker = folium.CircleMarker(
        location=[row['latitude'], row['longitude']],
        radius=5,  # Adjust the radius as needed
        color='blue',  # Border color
        fill=True,
        fill_color='blue',  # Fill color
        fill_opacity=0.7,  # Opacity of the fill
        tooltip=row['name'],  # Optional tooltip on hover
    )
    circle_marker.add_to(vdm_map) """

    # Add marker with number displayed inside using DivIcon
    folium.Marker(
        location=[row['latitude'], row['longitude']],
        icon=folium.DivIcon(
            html=f"""
                <div style="font-size: 12px; color: white; background-color: blue;
                            border-radius: 50%; width: 20px; height: 20px;
                            display: flex; justify-content: center; align-items: center;">
                    {index + 1}
                </div>
            """
        )
    ).add_to(vdm_map)




# Display the map
vdm_map
```
